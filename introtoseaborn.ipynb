```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('anur.csv')
print(df.head())
sns.barplot(x='Rank', y='Branch', data=df)
#print(df.columns)
plt.xlabel('Rank')
plt.ylabel('Branch')
plt.title('Bar Plot of Rank by Branch')


plt.show()
```
